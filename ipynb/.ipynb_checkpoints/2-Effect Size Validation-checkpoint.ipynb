# Introduction

We'll start by validating the ability of pseudo effect size, pseudo d, or $\widehat{\delta}$. We'll calculate this value from a test-based parametric power curve. We'll compare the fit value against an effect size calculated from an approximation of a one-sided z distribution (eq 1), and the optimizations solution of a t distribution or f distribution to compare performance.

We'll compare the $\widehat{\delta}$ to the test-based effect size ($d$ or $f^{2}$). We'll also compare the power calculated from the effect size to the power used to calculate the effect size.

The $\widehat{\delta}$ should reflect the test-based effect size predicted by Cohen for parameteric data, nad the power values should be able to round trip.

# Effect sizes

We're going to start by looking at the relationship between the effect size calculated using the test-based method and the effect size extrapolated using the z distribution.

The extrapolated effect size is calculated using the z distribution. The z distribution tests the alternative hypotheses,

$\begin{matrix}
\textbf{H}_{0}: & \mu_{1} = \mu_{2}\\
\textbf{H}_{1}: & \mu_{1} \neq \mu_{2}
\end{matrix}\tag{2.1}$

The hypothesis is tested using the Z test. The test statistic, $z$, is calculated according to equation (2.2) and drawn from a normal distribution with a mean of 0 and a standard deviation 1.
$\begin{align}
z = \sqrt{\frac{n}{2}}\frac{\left (\mu_{1} - \mu_{2} \right )}{\sigma}
\end{align}\tag{2.2}$

If the alternative hypothesis is true, the test statistic is drawn from a noncentral Z distribution ($Z'$) with mean, $\lambda$ and standard deviation, 1.
$\begin{align}
\lambda = \sqrt{\frac{n}{2}}\frac{\left ( \mu_{1} - \mu_{2} \right )}{\sigma}
\end{align}\tag{2.3}$

Power is calculated as

$\begin{align}
PWR &= 1 - \beta\\
&= 1 - \left ( P[Z' < z_{1 - \alpha / 2}] - P[Z' < z_{\alpha / 2}]\right )\\
&= 1 - \Phi_{Z'} \left(z_{1 - \alpha / 2}, \lambda \right ) + \Phi_{Z'} \left(z_{\alpha / 2}, \lambda \right )\\
&= 1 - \Phi_{Z'} \left (\lambda_{z} - z_{1 - \alpha/2} \right ) + \Phi_{Z} \left (z_{\alpha / 2} - \lambda_{Z} \right )\\
&= 1 - \Phi_{Z'} \left (\sqrt{\lambda_{z}^{2}} - z_{1 - \alpha/2} \right )\\
\end{align}\tag{2.4}$

We can re-arrange equation (2.4), using the cumulative distribution.
$\begin{align}
z_{1 - \beta} &= \sqrt{\lambda_{z}^{2}} - z_{1 - \alpha/2}\\
z_{1 - \beta} + z_{1 - \alpha/2} &= \sqrt{\lambda_{z}^{2}}\\
&= \sqrt{\frac{n}{2}\left (\frac{\left ( \mu_{1} - \mu_{2} \right )}{\sigma} \right )^{2}}
\end{align}\tag{2.5}$

Cohen defined the standard effect size, $d$, for a Z distribution as
$\begin{align}
d_{z} &= \frac{\mu_{1} - \mu_{2}}{\sigma}\\
&= \sqrt{\frac{2}{n}}\lambda_{z}\\
\end{align}\tag{2.6}$
We can substitute $d_{z}$ into equation (2.5), which lets us approximate the effect size as a fucntion of power and the number of observations.

$\begin{align}
z_{1 - \beta} + z_{1 - \alpha/2} &= \sqrt{\frac{n}{2}d_{z}^{2}}\\
d_{z} &= \sqrt{\frac{2\left (z_{1 - \beta} + z_{1 - \alpha/2} \right )^{2}}{n}}
\end{align}\tag{2.7}$

This makes the standard pseudo effect size, $\widehat{\delta}$ a function of the emperical power, $\widehat{PWR}$, according to equation (2.8).
$\begin{align}
\widehat{\delta} &= \sqrt{\frac{2\left (z_{\widehat{PWR}} + z_{1 - \alpha/2} \right )^{2}}{n}}
\end{align}\tag{2.8}$

The formula for this has been implemented in the `util` library which accompanies this notebook.

```python
for name, results in distributions.iteritems():
    for run_id, run in results.iteritems():
        if run_id == 'clean_name':
            continue
        test_power = run['test_power']
        effect = np.mean(ap.z_effect(counts, test_power))
        extr_power = ap.z_power(counts, effect)
        run['test_effect'] = effect
        run['extrapolated_power'] = extr_power
        run['skip'] = ~((test_power < 0.2).all() | (test_power > 0.8).all())
```

# Effect Size

We can start by comparing the effect size calculated using Cohen's methods with the effect size calculated from the test-based power.

```python
def_eff = []
sol_eff = []
skips = []
labels = []
summary1 = {}
summary2 = {}
for idx, name in enumerate(['anova', 'ttest_1', 'ttest_ind']):
    # Gets the effec tsizes 
    results = distributions[name]
    effects = [(run['base_effect'], run['test_effect'], run['skip'])
               for run in results.itervalues() if not isinstance(run, str)]
    x, y, i = zip(*effects)
    x = np.array(x)
    y = np.array(y)
    i = np.array(i)
    def_eff.append(x)
    sol_eff.append(y)
    skips.append(np.array(i))
    labels.append(results['clean_name'])
    
    # Calculates the correlation and ratio for all results
    ratio_a, bound_a = ap.ratio_error(x, y)
    r2_a = scipy.stats.pearsonr(x, y)[0]
    
    ratio_s, bound_s = ap.ratio_error(x[i], y[i])
    r2_s = scipy.stats.pearsonr(x[i], y[i])[0]
    
    # Summarizes the data
    summary1[results['clean_name']] = {'correlation': np.round(r2_a, 3),
                                       'ratio': np.round(ratio_a, 2),
                                       '95% CI': '[%1.2f, %1.2f]' 
                                       % (ratio_a - bound_a, ratio_a + bound_a)
                                       }
    summary2[results['clean_name']] = {'correlation': np.round(r2_s, 3),
                                       'ratio': np.round(ratio_s, 2),
                                       '95% CI': '[%1.2f, %1.2f]' 
                                       % (ratio_a - bound_s, ratio_a + bound_s)
                                       }
```

```python
ax = plot.plot_effect_sizes(def_eff, sol_eff,
                            xlabel="Cohen's d",
                            ylabel="Pseudo d"
                            )
fig = ax.figure
ax2 = fig.add_axes((0.6, 0.25, 0.3, 0.3))
plot.plot_effect_sizes(def_eff, sol_eff, ax=ax2)
ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_xlim([0, 1.5])
ax2.set_ylim([0, 1.5])
ax.legend(labels, loc=2)
fig.set_size_inches((4, 4))

pd.DataFrame.from_dict(summary1, orient='index')[['correlation', 'ratio', '95% CI']]
```

We can try excluding effect sizes associated with power less than 0.1 or greater than 0.8, to see if this improves the correlation.

```python
ax = plot.plot_effect_sizes([x[s] for x, s in zip(*(def_eff, skips))],
                            [y[s] for y, s in zip(*(sol_eff, skips))],
                            xlabel="Cohen's d",
                            ylabel="Pseudo d",
                            ticks=np.arange(0, 2.1, 0.5)
                            )
ax.legend(labels, loc=4)
fig = ax.figure
fig.set_size_inches((4, 4))
pd.DataFrame.from_dict(summary2, orient='index')[['correlation', 'ratio', '95% CI']]
```

There is a constant relationship between the $\widehat{\delta}$ effect size and the test-based effect size defined by Cohen.

Our $\widehat{\delta}$ is based on the Z test for two means, and can be expressed as
$\begin{align}
d_{z} &= \sqrt{\frac{n}{2}}\lambda_{z}\\
\end{align}\tag{2.6}$

The relationship between the independent two sample t test effect size, $d_{t2}$ and $\lambda_{t2}$, is
$\begin{align}
d_{t2} = \sqrt{\frac{n}{2}\lambda_{t2}^{2}}
\end{align}\tag{2.9}$
which explains the close to 1-1 relationship between the $d_{t2}$ and $\widehat{\delta}_{t2}$.

In comparison, the effect size for a one-sample t test, $d_{t2}$ and $\lambda_{t1}$ is
$\begin{align}
d_{t1} = \sqrt{n\lambda_{t1}^{2}}
\end{align}\tag{2.10}$
and the effect size for a one way ANOVA where $N$ samples are distributed across $k$ groups, $f^{2}$ and the noncentrality parameter, $\lambda_{f}$ is

$\begin{align}
f^{2} &= \sum_{i=1}^{k}{\left (\sqrt{\frac{1}{n_{i}}}\lambda_{i} \right )^{2}}\\
&= \sqrt{\frac{1}{N}}\sum_{i=1}^{k}{\lambda_{i}^{2}}
\end{align}\tag{2.11}$

This suggest a constant relationship of $\sqrt{2}$ between $d_{t1}$ and $\widehat{\delta}$ and a relationship of 2 between $f^{2}$ and $\widehat{\delta}$.

# Power comparisons

Next, we're going to compare the power we calculated using the noncentrality parameter with the power we extrapolate using the pseudo d effect size.

```python
color = '#1f78b4'

fig1, axes = plt.subplots(2, 4)
fig1.set_size_inches((8, 4))

xx = []
yy = []
summary1 = {}

for idx, name in enumerate(distributions.iterkeys()):
    # Gets the power vectors
    results = distributions[name]
    effects = [(run['test_power'],run['extrapolated_power'], run['skip'])
               for run in results.itervalues() if not isinstance(run, str)]
    x, y, i = zip(*effects)
    x = np.hstack(x)
    y = np.hstack(y)
    i = np.hstack([np.ones(counts.shape) * ii for ii in i])
        
    # Adds the data to the watch matrices
    xx.append(x)
    yy.append(y)
    
    plot.plot_power_comparison(x, y, 
                               axes[0, idx], 
                               axes[1, idx], 
                               color=color,
                               show_label=(idx == 0), 
                               ylabel1='Extrapolated Power',
                               ylabel2='Relative Difference',
                               title=results['clean_name'],
                               )
    ratio, bound = ap.ratio_error(x, y)
    r2 = scipy.stats.pearsonr(x, y)[0]
    round_ = int(-np.floor(np.log10(bound)))
    summary1[results['clean_name']] = {'correlation': np.round(r2, 3),
                                       'ratio': np.round(ratio, round_),
                                       'ratio_bound': np.round(bound, round_),
                                       }
pd.DataFrame.from_dict(summary1).transpose()
```

Again, it looks like low power values lead to errors. So, we'll exclude power values less than 0.1 from the original set.

```python
fig2, axes = plt.subplots(2, 4)
fig2.set_size_inches((8, 4))

xx = []
yy = []
summary1 = {}

for idx, name in enumerate(distributions.iterkeys()):
    # Gets the power vectors
    results = distributions[name]
    effects = [(run['test_power'],run['extrapolated_power'], run['skip'])
               for run in results.itervalues() if not isinstance(run, str)]
    x, y, i = zip(*effects)
    i = np.hstack([(np.ones(counts.shape) * ii).astype(bool) for ii in i])
    x = np.hstack(x)[i]
    y = np.hstack(y)[i]
    
        
    # Adds the data to the watch matrices
    xx.append(x)
    yy.append(y)
    
    plot.plot_power_comparison(x, y, 
                               axes[0, idx], 
                               axes[1, idx], 
                               color=color,
                               show_label=(idx == 0), 
                               ylabel1='Extrapolated Power',
                               ylabel2='Relative Difference',
                               title=results['clean_name'],
                               )
    ratio, bound = ap.ratio_error(x, y)
    r2 = scipy.stats.pearsonr(x, y)[0]
    round_ = int(-np.floor(np.log10(bound)))
    summary1[results['clean_name']] = {'correlation': np.round(r2, 3),
                                       'ratio': np.round(ratio, round_),
                                       'ratio_bound': np.round(bound, round_),
                                       }
pd.DataFrame.from_dict(summary1).transpose()
```
